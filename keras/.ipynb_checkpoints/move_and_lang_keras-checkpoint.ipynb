```python
import json,pickle
import re
saleVal = re.compile('[:.0-9a-z_A-Z!?%()|]')
def clean_sentence(text:str) -> str:
    """
    """
    #text = re.sub(r"<笑>",'',text)
    #text = "(D_ウズ)でもそれ鑑賞の一部だからしようと思っているんだけど:"
    text = saleVal.sub("",text)
    text = re.sub(r"(\()(.*?)\)",'',text)
    text = re.sub(r"<.?>",'',text)
    return text

usr2id = {'A:':0, 'B:':1, 'C:':2}
label2id = {'s':0, 'b':1, 'f':2, 'r':3, 'i':4}

def get_feature(text_files :list ,id_list :list) -> list:
    """
    """
    sentences = []
    label = []
    users = []
    for i in id_list:
        try:
            with open(text_files[i],encoding='shift-jis') as f:
                lines = f.readlines()
        except:
            with open(text_files[i],encoding='utf-8') as f:
                lines = f.readlines()

        lines = [line.rstrip().split() for line in lines]
        for line in lines:
            #if line[-1] == 'f':continue
            text = clean_sentence(line[-2])
            if text == "" or text in ["<笑>","<息>" ,"<声>"]:continue
            label.append(label2id[line[-1]])
            sentences.append(text)
            users.append(line[-3])

    users = get_usrID(users)

    return sentences, np.array(users,dtype=np.float32), np.array(label,dtype=np.int32)

def get_usrID(users :list) -> list:
    """ 
    直前の発話に対して，
        発話者が変わった場合 1
        発話者が同じ場合は 0
    :param: users : list 各発話の発話者(A,B,C)
    return: userID : list 発話者の変化(0,1)
    """
    userID = []
    #for i in range(len(users)-1):
    #if users[i] == users[i+1]:
    #userID.append(0) if users[i] == users[i+1] else userID.append(1)
    userID = [0 if users[i] == users[i+1] else 1 for i in range(len(users)-1)]
    #else:
    #userID.append(1)
    return userID

def wakati(sentences :list) -> list:
    """
    分かち書きして返す関数
    param: [私は学生です,私は犬です]
    return: [[私 は 学生 です],[私 は 犬 です]]
    """
    texts = []
    m = MeCab.Tagger("-d /usr/lib/mecab/dic/mecab-ipadic-neologd")
    for sentence in sentences:
        word_list = list()
        words_chasen = m.parse(sentence).split('\n')
        for word_chasen in words_chasen:
            if word_chasen == 'EOS': break
            #word = word_chasen.split("\t")[0]
            word_list.append(word_chasen.split("\t")[0])
        #word_list = ' '.join(word_list)
        texts.append(' '.join(word_list))
    return texts
```

```python
def make_Embedding(model, tokenizer, EMBEDDING_DIM=300) -> list:
    """
    重み行列作成
    """
    embedding_metrix = np.zeros((len(tokenizer.word_index)+1,EMBEDDING_DIM))
    for word, i in tokenizer.word_index.items():
        try:
            embedding_vector = model[word]
            embedding_metrix[i] = embedding_vector
        except KeyError:#学習済みモデルにない単語は，乱数で値を決定
            print(word)
            embedding_metrix[i] = np.random.normal(0, np.sqrt(0.25), EMBEDDING_DIM)
    return embedding_metrix

def save_word2id(tokenizer, INPUT=''):
    """
    単語→idの辞書を保存
    """
    import json
    f = open(INPUT,'w')#.json
    json.dump(tokenizer.word_index,f)
    f.close()

def save_EmbeddingMetrix(embedding_metrix : dict, INPUT : str = ''):
    """
    分散行列の保存@python3
    """
    
    f = open(INPUT,"wb")#
    pickle.dump(embedding_metrix,f,protocol=2)
    f.close()


def word2id(x :list ,MAX_SIZE :int = 10) -> list:
    """
    param: x...分かち書きされた１文章
    return: t...分かち書きされた各単語をidに変換したもの
    """
    t = np.zeros(MAX_SIZE)#[]
    for i,word in enumerate(x[-MAX_SIZE:]):
        word = word.lower()
        
        if word in filters: t[i] = 0
        else:
            try:
                t[i] = word_index[word]
            except:
                print(word,end=' ')
                t[i] = 0
    t = np.array(t).astype('int32')
    return t
```

```python
from keras.layers import *
import keras
import keras.backend as K
from keras_self_attention import SeqSelfAttention

import tensorflow as tf
from keras.backend.tensorflow_backend import set_session
config = tf.ConfigProto(
    gpu_options=tf.GPUOptions(
        visible_device_list="0", # specify GPU number
        allow_growth=True
    )
)
set_session(tf.Session(config=config))

def m(i1 = Input(shape=(10,)),i2 = Input(shape=(10,)),u_id = Input(shape=(1,))):
    
    Embed = keras.layers.Embedding((embedding_metrix.shape[0]), 300,mask_zero = True,weights = [embedding_metrix],input_length=10,trainable=False)
    x_a = Embed(i1)
    x_b = Embed(i2)
    #share or not
    shareLSTM = LSTM(64,return_sequences=True,dropout=0.5,recurrent_dropout=0.5,name='lstm')
    shareSelfAtt = SeqSelfAttention(units=32,name='self-att')
    shareSum = Lambda(lambda a: K.sum(a,axis=-2), output_shape=(64,))
    shareDense = Dense(16,activation='relu',name='dense')
    shareDropout = Dropout(0.5)
    for f in [shareLSTM,shareSelfAtt,shareSum,shareDense,shareDropout]:
        #for f in [shareLSTM,shareDense,shareDropout]:
        x_a = f(x_a)
        x_b = f(x_b)
    x = concatenate([x_a,x_b,u_id],axis=-1)
    o = Dense(5,activation='softmax')(x)
    m = keras.Model([i1,i2,u_id],o)
    m.compile(loss='sparse_categorical_crossentropy',metrics=['accuracy'],
              optimizer=keras.optimizers.Adam(1e-03)
              #optimizer=keras.optimizers.SGD(1e-03)
             )
    #m.summary()
    return m
```

```python
#attention
def m_att(i1 = Input(shape=(10,)),i2 = Input(shape=(10,)),u_id = Input(shape=(1,))):
    
    Embed = keras.layers.Embedding((embedding_metrix.shape[0]), 300,mask_zero = 0,weights = [embedding_metrix],input_length=10,trainable=False)
    x_a = Embed(i1)
    x_b = Embed(i2)
    #share or not
    x_a = LSTM(64,return_sequences=True,dropout=0.5,recurrent_dropout=0.5,name='lstm')(x_a)
    x = Dense(1,activation='linear')(x_a)
    x = Reshape((-1,))(x)
    x = Activation('softmax')(x)
    x = RepeatVector(64)(x)
    x = Permute((2,1))(x)
    x = Multiply()([x,x_a])
    x_a = Lambda(lambda a: K.sum(a,axis=-2), output_shape=(64,))(x)
    x_b = LSTM(64,return_sequences=True,dropout=0.5,recurrent_dropout=0.5,name='lstm1')(x_b)
    x = Dense(1,activation='linear')(x_b)
    x = Reshape((-1,))(x)
    x = Activation('softmax')(x)
    x = RepeatVector(64)(x)
    x = Permute((2,1))(x)
    x = Multiply()([x,x_b])
    x_b = Lambda(lambda a: K.sum(a,axis=-2), output_shape=(64,))(x)
    x = concatenate([x_a,x_b,u_id],axis=-1)
    o = Dense(5,activation='softmax')(x)
    m = keras.Model([i1,i2,u_id],o)
    m.compile(loss='sparse_categorical_crossentropy',metrics=['accuracy'],optimizer=keras.optimizers.Adam(1e-04))
    m.summary()
    return m

from sklearn.metrics import confusion_matrix
import seaborn as sns
class_weight = {0:1.0, 1:1.0, 2:6.0, 3:.0, 4:3.0}
for i in range(5):
    class_weight[i] = min (len(y) / Counter(y)[i], 15.0 )
print("class_weight",class_weight)
Acc=[]
for i in range(1):
    X1_train, X1_val, y_train, y_val = train_test_split(X1,y,test_size=0.2,random_state=i*2+1)
    X2_train, X2_val, y_train, y_val = train_test_split(X2,y,test_size=0.2,random_state=i*2+1)
    X_id_train, X_id_val, y_train, y_val = train_test_split(X_id,y,test_size=0.2,random_state=i*2+1)
    model = m_att()
    hist = model.fit(
                    [X1_train,X2_train,X_id_train],
                     y_train,
                     epochs=200,batch_size=128,
                     validation_data=([X1_val,X2_val,X_id_val],y_val),
                     callbacks = [keras.callbacks.EarlyStopping(monitor='val_loss',patience=10,verbose=1)],
                     verbose=2,
                     class_weight = class_weight
                    )
    #print(model.evaluate([X1_val,X2_val,X_id_val],y_val))
    acc = model.evaluate([X1_val,X2_val,X_id_val],y_val)[1]
    Acc.append(acc)
print(np.average(Acc))
pred = model.predict([X1_val,X2_val,X_id_val])
pred = [np.argmax(i) for i in pred]
print(confusion_matrix(y_val,pred))
```

```python
def add_positive_label(y):
    y_after = [0]*len(y)
    for i in range(len(y)):
        if y[i] != 1: continue
        for j in range(max(i-5,0),min(len(y),i+5)):
            y_after[j] = 1
    return y_after
```

```python
def lang_passive(i1 = Input(shape=(10,)),i2 = Input(shape=(10,)),u1_id = Input(shape=(1,)),u2_id = Input(shape=(1,))):
    
    Embed = keras.layers.Embedding((embedding_metrix.shape[0]), 300,mask_zero = True,weights = [embedding_metrix],input_length=10,trainable=False)
    x_a = Embed(i1)
    x_b = Embed(i2)
    #share or not
    shareLSTM = LSTM(64,return_sequences=False,dropout=0.5,recurrent_dropout=0.5,name='lstm')
    shareSelfAtt = SeqSelfAttention(units=32,name='self-att')
    shareSum = Lambda(lambda a: K.sum(a,axis=-2), output_shape=(64,))
    shareDense = Dense(16,activation='relu',name='dense')
    shareDropout = Dropout(0.5)
    for f in [shareSelfAtt,shareLSTM,shareDense,shareDropout]:
        x_a = f(x_a)
        x_b = f(x_b)
    x = concatenate([x_a,x_b,u1_id,u2_id],axis=-1)
    o = Dense(1,activation='sigmoid')(x)
    m = keras.Model([i1,i2,u1_id,u2_id],o)
    m.compile(loss='binary_crossentropy',metrics=['accuracy'],optimizer=keras.optimizers.Adam(1e-04))
    m.summary()
    return m
```
